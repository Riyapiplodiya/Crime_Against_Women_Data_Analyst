```python
x = df['Trafficking']  # This will be plotted on the X-axis
y = df['State/UT']  # This will be plotted on the Y-axis
# Step 3: Plot the data
plt.figure(figsize=(10,10))
plt.plot(x, y,marker='o', color='blue')

# Adding labels and title
plt.xlabel('Trafficking')
plt.ylabel('State/UT')
plt.title('Trafficking Count Over the States')

# Display the graph
plt.show()
```

```python
x = df['Dowry Deaths']  # This will be plotted on the X-axis
y = df['State/UT']  # This will be plotted on the Y-axis
# Step 3: Plot the data
plt.figure(figsize=(10,10))
plt.plot(x, y,marker='o', color='red')

# Adding labels and title
plt.xlabel('Dowry Deaths')
plt.ylabel('State/UT')
plt.title('Dowry Deaths Count Over the States')

# Display the graph
plt.show()
```

```python
x = df['Rape']  # This will be plotted on the X-axis
y = df['State/UT']  # This will be plotted on the Y-axis
# Step 3: Plot the data
plt.figure(figsize=(20,30))
plt.pie(x,labels=y,autopct='%1.1f%%')

# Adding labels and title
plt.xlabel('Rape')
plt.ylabel('State/UT')
plt.title('Rape, Count Over the States')

# Display the graph
plt.show()
```

```python
x = df['State/UT']  # This will be plotted on the X-axis
y = df['Acid Attack']  # This will be plotted on the Y-axis
# Step 3: Plot the data
plt.figure(figsize=(22,10))
plt.barh(x,y,color='green')

# Adding labels and title
plt.xlabel('State/UT')
plt.ylabel('Acid Attack')
plt.title('Acid Attack Over the States')

# Display the graph
plt.show()
```

```python
x = df['State/UT']  # This will be plotted on the X-axis
y = df['Domestic violence']  # This will be plotted on the Y-axis
# Step 3: Plot the data
plt.figure(figsize=(20,10))
plt.barh(x,y,color='pink')

# Adding labels and title
plt.xlabel('State/UT')
plt.ylabel('Domestic violence')
plt.title('Domestic violence Over the States')

# Display the graph
plt.show()
```

```python
x = df['Kidnapping/Abduction']  # This will be plotted on the X-axis
y = df['State/UT']  # This will be plotted on the Y-axis
# Step 3: Plot the data
plt.figure(figsize=(30,10))
plt.scatter(x,y,c='purple',marker='o')
# Adding labels and title
plt.xlabel('Kidnapping/Abduction')
plt.ylabel('State/UT')
plt.title('Kidnapping/Abduction Over the States')

# Display the graph
plt.show()
```
